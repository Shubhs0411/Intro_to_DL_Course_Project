```python
import pandas as pd
import os
import shutil
from PIL import Image, ImageDraw
```

```python
df = pd.read_csv('socal.csv')
```

```python
df.columns = ['frame', 'x1', 'y1', 'x2', 'y2', 'label']
```

```python
labels = ['cottonoid', 'drill', 'grasper', 'muscle', 'string', 'suction']
result_df = pd.DataFrame()

for label in labels:
    label_df = df[df['label'] == label].sample(n=200, random_state=42)
    result_df = pd.concat([result_df, label_df])

result_df = result_df.reset_index(drop=True)
```

```python
source_folder = 'JPEGImages/'
destination_folder = 'imagesv2/'
```

```python
os.makedirs(destination_folder, exist_ok=True)
```

```python
for image in jpeg_images:
    source_path = os.path.join(source_folder, image)
    destination_path = os.path.join(destination_folder, image)
    
    if os.path.exists(source_path):
        shutil.copy(source_path, destination_path)
        print(f"Copied: {image}")
    else:
        print(f"File not found: {image}")
```

```python
class_mapping = {
    'cottonoid': 0,
    'drill': 1,
    'grasper': 2,
    'muscle': 3,
    'scalpel': 4,
    'string': 5,
    'suction': 6,
    'tool': 7
}
```

```python
def convert_to_yolo_format(image_width, image_height, x1, y1, x2, y2):
    bbox_width = x2 - x1
    bbox_height = y2 - y1
    
    center_x = x1 + bbox_width / 2
    center_y = y1 + bbox_height / 2
    
    center_x /= image_width
    center_y /= image_height
    bbox_width /= image_width
    bbox_height /= image_height
    return center_x, center_y, bbox_width, bbox_height
```

```python
labels_dir = 'labelsv2'
os.makedirs(labels_dir, exist_ok=True)
```

```python
filtered_df = df[df['frame'].isin(result_df['frame'])]
```

```python
filtered_df['label'].value_counts()
```

```python
len(filtered_df)
```

```python
image_folder = 'imagesv2'
for frame in filtered_df['frame'].unique():
    frame_df = filtered_df[filtered_df['frame'] == frame]
    image_path = os.path.join(image_folder, frame)
    
    try:
        with Image.open(image_path) as img:
            image_width, image_height = img.size
    except Exception as e:
        print(f"Error Opening the image {frame}: {e}")
        continue
        
    yolo_data = []
    for _, row in frame_df.iterrows():
        class_id = class_mapping.get(row['label'], -1)
        if class_id == -1:
            continue
        
        center_x, center_y, bbox_width, bbox_height = convert_to_yolo_format(image_width, image_height, row['x1'], row['y1'], row['x2'], row['y2'])
        
        yolo_data.append(f"{class_id} {center_x} {center_y} {bbox_width} {bbox_height}")
        
    file_name = os.path.splitext(frame)[0] + ".txt"
    file_path = os.path.join(labels_dir, file_name)
    
    with open(file_path, 'w') as f:
        f.write("\n".join(yolo_data))

print("YOLO format conversion is complete.....")   
```

```python
import random

images_folder = 'imagesv2/'
labels_folder = 'labelsv2/'
train_images_folder = 'imagesv2/train/'
test_images_folder = 'imagesv2/test/'
train_labels_folder = 'labelsv2/train/'
test_labels_folder = 'labelsv2/test/'

os.makedirs(train_images_folder, exist_ok=True)
os.makedirs(test_images_folder, exist_ok=True)
os.makedirs(train_labels_folder, exist_ok=True)
os.makedirs(test_labels_folder, exist_ok=True)
```

```python
image_files = os.listdir(images_folder)

image_files = [f for f in image_files if f.endswith('.jpeg')]

random.shuffle(image_files)

train_images = image_files[:960]
test_images = image_files[960:1200] 
```

```python
for image in train_images:
    image_src = os.path.join(images_folder, image)
    label_src = os.path.join(labels_folder, image.replace('.jpeg', '.txt')) 
    image_dst = os.path.join(train_images_folder, image)
    label_dst = os.path.join(train_labels_folder, image.replace('.jpeg', '.txt'))

    if os.path.exists(image_src) and os.path.exists(label_src):
        shutil.move(image_src, image_dst)
        shutil.move(label_src, label_dst)
        print(f"Moved to train: {image}, {image.replace('.jpeg', '.txt')}")
    else:
        print(f"File missing: {image_src} or {label_src}")
```

```python
for image in test_images:  
    image_src = os.path.join(images_folder, image)
    label_src = os.path.join(labels_folder, image.replace('.jpeg', '.txt'))

    image_dst = os.path.join(test_images_folder, image)
    label_dst = os.path.join(test_labels_folder, image.replace('.jpeg', '.txt'))

    if os.path.exists(image_src) and os.path.exists(label_src):
        shutil.move(image_src, image_dst)
        shutil.move(label_src, label_dst)
        print(f"Moved to test: {image}, {image.replace('.jpeg', '.txt')}")
    else:
        print(f"File missing: {image_src} or {label_src}")
```
